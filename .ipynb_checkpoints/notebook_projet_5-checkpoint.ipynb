## Etape 1 - Faites la migration vers MongoDB

### 1.1. Définition des concepts clés de MongoDB

**Document**
Un document est l'unité de base des données, constituée de paires champ-valeur. Les documents peuvent contenir des chaînes de caractères, des nombres, des dates, des tableaux et même d'autres documents intégrés. Cela permet de créer des modèles de données sophistiqués représentant des relations complexes dans un seul document, ce qui correspond à la façon dont les objets sont structurés dans la plupart des langages de programmation. <br>
Il contient les données sous un format proche du JSON appelé BSON (Binary JSON).


**Collection**
Une collection regroupe des documents apparentés, à l'instar d'un tableau mais avec plus de souplesse. Les collections n'imposent pas de schéma, de sorte que les documents peuvent avoir des structures et des champs différents. Les collections existent dans les bases de données MongoDB et chaque base de données peut contenir plusieurs collections.


**Base de données**
Une base de données MongoDB contient plusieurs collections.

source: https://www.datacamp.com/fr/blog/what-is-mongodb

## Étape 2 - Conteneurisez l'application avec Docker

#### 2.1. Docker-compose.yml

Source: https://docs.docker.com/reference/compose-file/

**Structuration du fichier docker-compose.yml**

```
services:
  # Service MongoDB  
  mongo_datasolutech:  
    image: mongo:7 #Docker télécharge l’image officielle MongoDB  https://hub.docker.com/_/mongo  
    container_name: mongodb_datasolutech #On donne un nom clair au conteneur MongoDB    
    volumes:    
      - volume_mongodb:/data/db   # Les données MongoDB sont sauvegardées dans un volume Docker      
    ports:    
      - "27018:27017" #Cela rend MongoDB accessible depuis sur le port 27017, voir schéma ci-dessous
      
   # Service de test
  test:
    #profiles:
     #- test
    env_file:
      - .env     
    build: 
      context: . # le . signifie le dossier courant où se trouve le docker_compose.
      dockerfile: Dockerfile.test # Docker construit l’image du script Python à partir du Dockerfile.test
    container_name: test_datasolutech
    depends_on:
      - mongo_datasolutech # Cela indique que le conteneur mongo doit être démarré avant le conteneur migration
    volumes:
      - ./donnees:/data #Le dossier local donnees est partagé avec le conteneur Python

  # Service de migration  
  migration:  
    env_file:    
      - .env  #On spécifie que l'environnement est le .env      
    build: . # Docker construit l’image du script Python à partir du Dockerfile    
    container_name: migration_datasolutech    
    depends_on:    
      - mongo_datasolutech # Cela indique que le conteneur mongo doit être démarré avant le conteneur migration      
    volumes:    
      - ./donnees:/data #Le dossier local donnees est partagé avec le conteneur Python  
volumes:  
  volume_mongodb: # volume pour les données MongoDB

```
